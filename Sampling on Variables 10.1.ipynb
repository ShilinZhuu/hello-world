### Function 1 for test data

```python
#correct for rating data, one of the final version, 1 is at the already downgraded time period
#unsolved corner case,only marked for the same id  
dftest['Comparison']=0
def Row_Comparetest (col_name):
    for rows in range(1,len(dftest)-1):
        if (not pd.isnull(dftest.at[rows,col_name])| pd.isnull(dftest.at[rows-1,col_name])):
            diff = dftest.at[rows,col_name] - dftest.at[rows-1,col_name]
            if diff > 0:  ##Downgrade cases
                dftest.at[rows,'Comparison']=1   ##'colname will not work'
```
